```python
def create_experience_database(trips):
    exp_tags = []
    exp_ids = []
    exp_dates = []
    genders = []
    ages = []
    weights = []
    authors = []
    views = []
    drug_list = []
    pub_date = []
    for trip in trips:
        if trip:
            exp_tags.append(get_exptags(trip))
            exp_ids.append(get_expid(trip))
            exp_dates.append(get_expdate(trip))
            genders.append(get_gender(trip))
            ages.append(get_age(trip))
            weights.append(get_bodyweight(trip))
            authors.append(get_author(trip))
            views.append(get_views(trip))
            dosechart = get_dosechart(trip)
            drug_list.append(get_unique_drugs(dosechart))
            pub_date.append(get_pubdate(trip))
    df = pd.DataFrame({'exp_id': exp_ids, 'tags':exp_tags, 'date':exp_dates,
                      'gender':genders, 'age':ages,'weight':weights,
                      'author':authors, 'views':views, 'substances':drug_list, 'pub_date':pub_date})
    return(df)
```

```python
def populate_drug_counts(trips, empty_drug_table):
    n = 0
    drug_table = empty_drug_table
    for trip in trips:
        if trip:
            dosechart = get_dosechart(trip)
            unique_drugs = get_unique_drugs(dosechart)
            cleaned_unique_drugs = []
            for i in unique_drugs:
                if str(i) == 'nan':
                    cleaned_unique_drugs.append('No Substance')
                else:
                    cleaned_unique_drugs.append(i)
            x = cleaned_unique_drugs
            y = cleaned_unique_drugs
            for drug in x:
                for combo in y:
                    drug_table[combo][drug] = drug_table[combo][drug] + 1 
        else:
            pass
        if n % 100 == 0:
            print(n)
        n += 1
    return(drug_table)
    
```

```python
def Eparse_base_nodes(name, drug_class):
    id = name.lower()
    group = get_drug_group(drug_class)
    label = name
    level = 2
    size = round(np.interp(drug_table[name][name], xp = [1, 250, 500], fp = [1, 25, 50]))
    return("{id: '%s', group: %i, label: '%s', level: %i, size: %i}," %(id, group, label, level, size))
```

```python
def Eparse_base_links(source, target, val):
    target_l = target.lower()
    source_l = source.lower()
    strength =  round(np.interp(val, xp = [1, 250, 500], fp = [0.01, 0.1, 0.6]), 4)
    return("{target: '%s', source: '%s', strength: %f}," %(target_l, source_l, strength))
```
